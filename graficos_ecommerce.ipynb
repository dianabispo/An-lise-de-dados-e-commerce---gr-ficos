```python
# Importando a biblioteca matplotlib para gráficos
import matplotlib.pyplot as plt

# Gráfico de histograma para 'Nota'
plt.figure(figsize=(10,6))
plt.hist(df_csv['Nota'], bins=20, color='skyblue', edgecolor='black')
plt.title('Distribuição das Notas')
plt.xlabel('Nota')
plt.ylabel('Frequência')
plt.show()
```

```python
# Gráfico de dispersão entre 'Nota' e 'N_Avaliações'
plt.figure(figsize=(10,6))
plt.scatter(df_csv['Nota'], df_csv['N_Avaliações'], color='purple')
plt.title('Relação entre Nota e Número de Avaliações')
plt.xlabel('Nota')
plt.ylabel('Número de Avaliações')
plt.show()
```

```python
# Importando as bibliotecas necessárias
import seaborn as sns
import matplotlib.pyplot as plt

# Calculando a correlação entre as variáveis numéricas
correlacao = df_csv.corr()

# Gráfico de mapa de calor de correlação
plt.figure(figsize=(10,6))
sns.heatmap(correlacao, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Mapa de Calor - Correlação entre Variáveis')
plt.show()

```

```python
# Gráfico de barra para a frequência dos 'Materiais'
plt.figure(figsize=(10,6))
df_csv['Material'].value_counts().plot(kind='bar', color='lightgreen')
plt.title('Frequência dos Materiais')
plt.xlabel('Material')
plt.ylabel('Frequência')
plt.show()
```

```python
# Gráfico de pizza para a distribuição de 'Gênero'
plt.figure(figsize=(8,8))
df_csv['Gênero'].value_counts().plot(kind='pie', autopct='%1.1f%%', colors=['lightblue', 'lightpink'])
plt.title('Distribuição de Gêneros')
plt.ylabel('')
plt.show()
```

```python
# Gráfico de densidade para 'Desconto'
plt.figure(figsize=(10,6))
df_csv['Desconto'].plot(kind='density', color='purple')
plt.title('Distribuição de Densidade do Desconto')
plt.xlabel('Desconto')
plt.ylabel('Densidade')
plt.show()
```

```python
# Gráfico de regressão entre 'Nota' e 'Preço_MinMax'
plt.figure(figsize=(10,6))
sns.regplot(x='Nota', y='Preço_MinMax', data=df_csv, scatter_kws={'color':'blue'}, line_kws={'color':'red'})
plt.title('Gráfico de Regressão - Nota vs Preço Mínimo Máximo')
plt.xlabel('Nota')
plt.ylabel('Preço Mínimo Máximo')
plt.show()
```
